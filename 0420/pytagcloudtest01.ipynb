```python
from collections import Counter
```

```python
a=['홍길동','홍길동','배선한','배선한','박경미']
Counter(a)
```

```python
f=open('data/yes24_review.txt','r')
data=f.read()
print(data)
```

```python
import re
data=re.sub(r'[^/w]','',data) #워드가 아닌글자 제거
```

```python
from konlpy.tag import Okt
import pytagcloud
```

```python
okt=Okt()
nouns=okt.nouns(data)
print(nouns)
```

```python
counter=Counter(nouns)
print(counter)
```

```python
counter2=counter.most_common(50)
print(counter2)
```

```python
wordInfo=dict()
for t,c in counter2:
    if (len(t)>1):
        wordInfo[t]=c
        # print(t,c)
# counter3=dict(wordInfo).items()
counter3=wordInfo.items()
```

```python
# taglist=pytagcloud.make_tags(counter2, maxsize=80)
taglist=pytagcloud.make_tags(counter3, maxsize=80)
taglist
```

```python
pytagcloud.create_tag_image(taglist,'data/pytagcloud02.png',size=(800,600),fontname='Korean')
```
