```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
import statsmodels.api as sm
```

```python
data=pd.read_csv('data/train-data.csv')
data
```

```python
data1=data.loc[:,['Name','Year', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'New_Price', 'Price']]
data1
```

```python
data1.New_Price=data1.New_Price.fillna('0')
data1
```

```python
new_price=list(data1.New_Price)
print(new_price)
price=list(data1.Price)
print(price)
```

```python
for i in range(len(new_price)):
    if new_price[i]!='0':
        price[i]=float(str(new_price[i]).split()[0])

data1.Price=price
data1
```

```python
data1.isna().sum()
```

```python
data1=data1.dropna()
data1.isna().sum()
```

```python
data1
```

```python
names=list(data1.Name)
mileage=list(data1.Mileage)
engine=list(data1.Engine)

for i in range(len(names)):
    names[i]=names[i].split()[0]
    mileage[i]=float(mileage[i].split()[0])
    engine[i]=float(engine[i].split()[0])
data1.Name=names
data1.Mileage=mileage
data1.Engine=engine

data1
```

```python
data1.info()
```

```python
data1=data1.drop(columns='New_Price')
data1
```

```python
from scipy import stats
from statsmodels.formula.api import ols, glm
```

```python
formula = 'Price ~ Year + Kilometers_Driven + Mileage + Engine'


```

```python
regression_result=ols(formula,data=data1).fit()
```

```python
regression_result.summary()
```

```python
sample=data1[data1.columns.difference(['Price'])]
sample1=sample[0:10][:]
sample1
```

```python
smaple_predict=regression_result.predict(sample1)
print(smaple_predict)
```
