```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```python
red_df=pd.read_csv('data/winequality-red.csv', sep=';')
red_df.to_csv('data/winequality-red2.csv', index=False)
```

```python
white_df=pd.read_csv('data/winequality-white.csv', sep=';')
white_df.to_csv('data/winequality-white2.csv', index=False)
```

```python
red_df.insert(0,column='type',value='red')
white_df.insert(0,column='type',value='white')
```

```python
print(red_df.head())
print(white_df.head())
```

```python
print(red_df.shape)
print(white_df.shape)
```

```python
wine=pd.concat([red_df,white_df])
print(wine.shape)
```

```python
wine.to_csv('data/wine.csv', index=False)
```

```python
wine.info()
```

```python
wine.columns=wine.columns.str.replace(' ','_')
print(wine.head())
```

```python
from scipy import stats
from statsmodels.formula.api import ols, glm
```

```python
red_wine_quality=wine.loc[wine['type']=='red','quality']
white_wine_quality=wine.loc[wine['type']=='white','quality']
```

```python
print(red_wine_quality)
print(white_wine_quality)
```

```python
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var=False)
```

```python
sns.pairplot(wine[['quality','fixed_acidity','residual_sugar','pH','alcohol']])
```

```python
formula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + \
      residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + \
      density + pH + sulphates + alcohol'

regression_result=ols(formula,data=wine).fit()

```

```python
regression_result.summary()
```

```python
sample=wine[wine.columns.difference(['quality','type'])]
sample1=sample[0:5][:]
sample1
```

```python
smaple_predict=regression_result.predict(sample1)
print(smaple_predict)
```
