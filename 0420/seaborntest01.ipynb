#1. countplot()
- 항목의 갯수를 세어줌
- 알아서 column을 구성하고 있는 value들을 구분하여 보여줌

```python
sns.countplot(x='class',data=titanic,hue='who')
plt.show()
```

```python
sns.countplot(y='class',data=titanic,hue='who')
plt.show()
```

```python
sns.countplot(x='class',data=titanic,hue='who', palette='Pastel1_r')
plt.show()
```

https://www.geeksforgeeks.org/seaborn-color-palette/

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
plt.show()
```

```python
sns.palplot(sns.color_palette("Greys"))
plt.show()
```

```python
sns.palplot(sns.color_palette("terrain_r", 7))
plt.show()
```

```python
color=['Blue','White','Red','Yellow','Green','Grey']
sns.set_palette('deep')
sns.countplot(x='class',hue='who',data=titanic)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='sex', size='size')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker',row='sex', col='day', data=tips)
plt.show()
```

2. taxis에서 total과 tip 사이의 상관관계 그래프를 작성하고, hue는 color를 지정하고 payment를 column 으로 지정하는 subplot 작성

```python

taxis=sns.load_dataset('taxis')

sns.relplot(data =taxis, x='total', y='tip', hue='color', col='payment' )
```
